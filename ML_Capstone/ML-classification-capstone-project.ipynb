# Grid Search

```python
X_train
```

```python
def grid_search(X_train, X_test, y_train, y_test, vectorizer, vectorizer_name, param_list):
    print(f"\n{'='*10} Using {vectorizer_name} vectorization {'='*10}\n")
    best_models = {}
    
    for name, model, param_grid in param_list:
        print(f"Training {name} with GridSearchCV using {vectorizer_name} vectorization")
        
        pipeline = Pipeline([
            (vectorizer_name, vectorizer),
            ('clf', model)
        ])
        
        grid = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='accuracy')
        grid.fit(X_train, y_train)
        best_models[name] = grid.best_estimator_
        
        print(f"Best parameters for {name}: {grid.best_params_}")
        print(f"Best CV accuracy for {name}: {grid.best_score_}")
        
        y_pred = grid.predict(X_test)
        print(f"\n{vectorizer_name} {name} Classification Report:")
        print(classification_report(y_test, y_pred), "\n")
        
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=grid.best_estimator_['clf'].classes_)
        disp.plot(cmap='Purples')
        plt.title(f"{vectorizer_name} - {name} Confusion Matrix")
        plt.show()
        
    return best_models

tfidf_vector = TfidfVectorizer(stop_words='english', ngram_range=(1,3))
count_vector = CountVectorizer(stop_words='english', ngram_range=(1,3))
```

# Gradient Boosting

```python
gb_params = {
    'clf__learning_rate': [0.01, 0.1, 1],
    'clf__n_estimators': [50, 100, 150]
}

gb_list = [
    ("GradientBoosting", GradientBoostingClassifier(), gb_params)
]

gb_tfidf = grid_search(X_train['text'], X_test['text'], y_train['class'], y_test['class'], tfidf_vector, "TF-IDF", gb_list)
```

```python
joblib.dump(gb_tfidf, "gb_tfidf.pkl")
```

```python
gb_pipeline = gb_tfidf['GradientBoosting']

gb_classifier = gb_pipeline.named_steps['clf']

importances = gb_classifier.feature_importances_
```

```python
feature_names = gb_pipeline.named_steps['TF-IDF'].get_feature_names_out()


features_importance = sorted(zip(feature_names, importances), key=lambda x: x[1], reverse=True)
for feature, importance in features_importance[:20]:
    print(f"{feature}: {importance:.4f}")
```
